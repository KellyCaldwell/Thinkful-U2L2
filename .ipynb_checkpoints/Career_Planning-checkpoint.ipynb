## Section 1: Explore the landscape
In the past, a lot of business decisions were made by the intuition of experienced leaders. That’s no longer a viable, competitive strategy. Intuition will not hold up in a match against designed,  and tested experiments. 

Data science is truly interdisiplinary. Though each individual career can be vastly different than the next, a data scientist is less likely to shy away from the diversity in favor of problem complexity. 

In the end, no matter the tools, approach, or background, a data scientist can uncover the otherwise unintelligable and deliver thoughtful models based on demostrable results.
